```python
# load libraries
import pandas as pd
```

```python
# load data
MO = pd.read_csv('foodinsecurity_MO.csv')
KS = pd.read_csv('foodinsecurity_KS.csv')
```

```python
MO.head()
```

```python
# add columns with states
MO['State'] = 'Missouri'
KS['State'] = 'Kansas'

# mark as part of KC or not, for future analysis purposes
# used official list of counties in Kansas City Metropolitan Statistical Area: 
# http://www.marc.org/Data-Economy/Metrodataline/General-Information/Statistical-Areas
MO['KC_or_no'] = 'No'
MO.loc[MO['county'].isin(['Cass','Clay','Clinton','Jackson','Lafayette',
                                 'Platte','Ray','Bates','Caldwell']), 'KC_or_no'] = 'Yes'
KS['KC_or_no'] = 'No'
KS.loc[KS['county'].isin(['Johnson','Leavenworth','Miami','Wyandotte',
                                  'Franklin','Linn']), 'KC_or_no'] = 'Yes'

# merge two files
insecure = pd.concat([MO,KS])
insecure = insecure.reset_index()
insecure.head()
```

Questions:
1. Which KC counties have the highest rate of child insecurity?
2. Which KC counties have the highest overall rate of insecurity?
3. Which KC counties have the highest rate of insecure kids who are likely ineligible for aid?

```python
# 1. Which KC counties have the highest rate of child insecurity?
by_child = insecure[['county','child_insecure_rate']][insecure.KC_or_no == 'Yes'].sort_values('child_insecure_rate', ascending=False)
by_child
```

Result: Wyandotte is above KS's statewide rate of 18.3% and ranks 19th out of all counties in the 2 states. Linn, Franklin, Ray, Bates, Miami, Leavenworth and Jackson also exceed both states' rates. Wyandotte and Jackson are part of the heart of KC. The other county that makes up KC -- Clay County, Missouri -- ranks below both statewide rates. 

```python
# 2. Which KC counties have the highest overall rate of insecurity?
by_insecure = insecure[['county','insecure_rate']][insecure.KC_or_no == 'Yes'].sort_values('insecure_rate', ascending=False)
by_insecure
```

Result: Jackson jumps to the top, followed by Wyandotte. These are the only two KC counties that exceed both statewide rates. In all KC counties, rates of child insecurity are higher than rates for all.

```python
# 3. Which KC counties have the highest rate of insecure kids who are likely ineligible for aid?
by_ineligible = insecure[['county','insecure_ineligible']][insecure.KC_or_no == 'Yes'].sort_values('insecure_ineligible', ascending=False)
by_ineligible
```

Result: Most of Wyandotte's kids are eligible for aid, the second highest rate in the metro area. Clay and Jackson perform worse, but more than half of the children in each of those counties is eligible, too.
